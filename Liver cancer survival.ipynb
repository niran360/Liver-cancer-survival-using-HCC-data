# Modelling

```python
X = df3.drop(['Class'], axis=1)
y = df3['Class']
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC, LinearSVC
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import fbeta_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report,confusion_matrix
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=0)


print("Before OverSampling, counts of label '1': {}".format(sum(y_train==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train==0)))
```

```python
sm = SMOTE(random_state=2)
X_train_sm, y_train_sm = sm.fit_sample(X_train, y_train.ravel())


print("After OverSampling, counts of label '1': {}".format(sum(y_train_sm==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_sm==0)))
```

```python
# X_train_sm
```

```python
# standardizing the data
sc = StandardScaler()
X_train_sm = sc.fit_transform(X_train_sm)
X_val = sc.transform(X_val)
```

```python
score_list = []
for each in range(1,20):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(X_train_sm,y_train_sm)
    y_head=knn2.predict(X_val)
    score_list.append(accuracy_score(y_val,y_head))
plt.figure(figsize=(8, 8))    
plt.plot(range(1,20),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.grid() 
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=1, leaf_size=10, p=1, weights='uniform')
svm_lin=SVC(kernel= 'linear', C=1, )
logr=LogisticRegression(C=0.1, solver='liblinear', max_iter=300, multi_class = 'ovr', tol=0.0001)
rf=RandomForestClassifier(warm_start=True, criterion = 'entropy',min_samples_split=3, min_samples_leaf=3, n_estimators=80, bootstrap=False )
mlp= MLPClassifier(hidden_layer_sizes=(50,250,50), max_iter=500, alpha=0.01,learning_rate='invscaling', max_fun=15000)
models=[knn,mlp,svm_lin,logr,rf]
```

```python
knn.fit(X_train_sm, y_train_sm)
knnpred=knn.predict(X_val)
svm_lin.fit(X_train_sm,y_train_sm)
svm_pred=svm_lin.predict(X_val)
```

```python
mlp.fit(X_train_sm,y_train_sm)
mlp_pred=mlp.predict(X_val)
logr.fit(X_train_sm,y_train_sm)
logpred=logr.predict(X_val)
rf.fit(X_train_sm,y_train_sm)
rfpred=rf.predict(X_val)
```

```python
accuracyknn=accuracy_score(y_val,knnpred)*100
f1scoreknn=f1_score(y_val,knnpred)*100
f2scoreknn=fbeta_score(y_val,knnpred, beta=2)*100
aucknn=roc_auc_score=(y_val,knnpred)
precisionknn=precision_score(y_val,knnpred)*100
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(confusion_matrix(y_val,knnpred),annot=True,fmt="d")
```

```python
accuracysvm=accuracy_score(y_val,svm_pred)*100
f1scoresvm=f1_score(y_val,svm_pred)*100
f2scoresvm=fbeta_score(y_val,svm_pred, beta=2)*100
aucsvm=roc_auc_score=(y_val,svm_pred)
precisionsvm=precision_score(y_val,svm_pred)*100
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(confusion_matrix(y_val,svm_pred),annot=True,fmt="d")
```

```python
accuracylogr=accuracy_score(y_val,logpred)*100
f1scoreslogr=f1_score(y_val,logpred)*100
f2scorelogr=fbeta_score(y_val,logpred, beta=2)*100
auclogr=roc_auc_score=(y_val,logpred)
precisionlogr=precision_score(y_val,logpred)*100
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(confusion_matrix(y_val,logpred),annot=True,fmt="d")
```

```python
accuracyrfpred=accuracy_score(y_val,rfpred)*100
f1scorerfpred=f1_score(y_val,rfpred)*100
f2scorerfpred=fbeta_score(y_val,rfpred, beta=2)*100
aucrfpred=roc_auc_score=(y_val,rfpred)
precisionrfpred=precision_score(y_val,rfpred)*100
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(confusion_matrix(y_val,rfpred),annot=True,fmt="d")
```

```python
accuracymlp=accuracy_score(y_val,mlp_pred)*100
f1scoremlp=f1_score(y_val,mlp_pred)*100
f2scoremlp=fbeta_score(y_val,mlp_pred, beta=2)*100
aucmlp=roc_auc_score=(y_val,mlp_pred)
precisionmlp=precision_score(y_val,mlp_pred)*100
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(confusion_matrix(y_val,mlp_pred),annot=True,fmt="d")
```

```python
plt.figure(figsize=(6, 4))
fitted_models, scores=ds.model.compare_model(models,X_train_sm,y_train_sm,  scoring_cv=5)
```

```python
from sklearn.model_selection import cross_val_score
knnscores = cross_val_score(knn, X_train_sm,y_train_sm, cv=5)
svm_linscores = cross_val_score(svm_lin, X_train_sm,y_train_sm, cv=5)
logrscores = cross_val_score(logr, X_train_sm,y_train_sm, cv=5)
rfscores = cross_val_score(rf, X_train_sm,y_train_sm, cv=5)
mlpscores = cross_val_score(mlp, X_train_sm,y_train_sm, cv=5)
```

```python
Iteration_1=[svm_linscores[0],svm_linscores[1],svm_linscores[2],svm_linscores[3],svm_linscores[4]]
Iteration_2=[logrscores[0],logrscores[1],logrscores[2],logrscores[3],logrscores[4]]
Iteration_3=[rfscores[0],rfscores[1],rfscores[2],rfscores[3],rfscores[4]]
Iteration_4=[knnscores[0],knnscores[1],knnscores[2],knnscores[3],knnscores[4]]
Iteration_5=[mlpscores[0],mlpscores[1],mlpscores[2],mlpscores[3],mlpscores[4]]
```

```python
crossval=pd.DataFrame()
index=['1','2','3','4','5']
Algorithm=["SVM","Logistic Regression","Random Forest","KNN","MLP Classifier"]
Iterations=["Iteration1","Iteration2","Iteration3","Iteration4","Iteration5"]
crossval["Iterations"]=index
# crossval["Algorithm"]=Algorithm
crossval["SVM"]=Iteration_1
crossval["Logistic Regression"]=Iteration_2
crossval["Randomm Forest"]=Iteration_3
crossval["KNN"]=Iteration_4
crossval["MLP Classifier"]=Iteration_5
crossval
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
ax = plt.gca()

crossval.plot(kind='line',x='Iterations',y='SVM',ax=ax)
crossval.plot(kind='line',x='Iterations',y='Logistic Regression', color='red', ax=ax)
crossval.plot(kind='line',x='Iterations',y='Randomm Forest', color='blue', ax=ax)
crossval.plot(kind='line',x='Iterations',y='KNN', color='cyan', ax=ax)
crossval.plot(kind='line',x='Iterations',y='MLP Classifier', color='yellow', ax=ax)
# plt.figure(figsize=(4, 4))
plt.show()
```

```python
accuracy=[accuracysvm,accuracylogr,accuracyrfpred,accuracyknn,accuracymlp]
f1=[f1scoresvm,f1scoreslogr,f1scorerfpred,f1scoreknn,f1scoremlp]
f2=[f2scoresvm,f2scorelogr,f2scorerfpred,f2scoreknn,f2scoremlp]
roc=[aucsvm,auclogr,aucrfpred,aucknn,aucmlp]
precise=[precisionsvm,precisionlogr,precisionrfpred,precisionknn,precisionmlp]
```

```python
Results=pd.DataFrame()
Algorithm=["SVM","Logistic Regression","Random Forest","KNN","MLP Classifier"]
Results["Algorithms"]=Algorithm
Results["Accuracy"]=accuracy
Results["F1Score"]=f1
Results["F2Score"]=f2
Results["Roc"]=roc
Results["Precision"]=precise
Results
```
